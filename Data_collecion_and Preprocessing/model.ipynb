# Preprocessing the data

```python
from nltk.corpus import stopwords
from nltk.stem import  PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
lemmatizer = WordNetLemmatizer()
stopw = set(stopwords.words("english"))
stopw.update(".", "?", "(", ")", ",", "-", "'", '"',"/","2","1","3","4","5","6","7","8","9","10","&")
```

```python
def text_to_words(raw_text):
    
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    lemmatizer.lemmatize(words)
    words = words.split()
    words = [word for word in words if not word.startswith('rt')]
    meaningful_words = [w for w in words if not w in stopw]
    
    return (" ".join(meaningful_words))
```

```python
final['clean_text'] = final['text'].apply(text_to_words)
final
```

```python
def country_name(raw_text):
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    words = words.split()
    for word in words:
        if word == 'spain'or word == 'Portugal' or word =='Italy' or word == 'Turkey' or word == 'Cyprus'or word == 'Africa':
                words = word
        else:
            words = 'Nan'
    return words
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.basemap import Basemap
```

```python
m = Basemap(resolution='c',
            projection='kav7',
            lat_0=0., # Center around
            lon_0=0.) # lat 0, lon 0
```

```python
n_graticules = 18
parallels = np.arange(-80., 90, n_graticules)
meridians = np.arange(0., 360., n_graticules)
lw = 1
dashes = [5,7] # 5 dots, 7 spaces... repeat
graticules_color = 'grey'
```

```python
fig1 = plt.figure(figsize=(16,20))
fig1.patch.set_facecolor('#e6e8ec')
ax = fig1.add_axes([0.1,0.1,0.8,0.8])

m.drawmapboundary(color='white', 
                  linewidth=0.0, 
                  fill_color='white')
m.drawparallels(parallels, 
                linewidth=lw, 
                dashes=dashes, 
                color=graticules_color)
m.drawmeridians(meridians, 
                linewidth=lw, 
                dashes=dashes, 
                color=graticules_color)
m.drawcoastlines(linewidth=0)
m.fillcontinents('black', 
                 lake_color='white')
m.drawcountries(linewidth=1, 
                linestyle='solid', 
                color='white', 
                zorder=30)

title = plt.title('World map (Kavrayskiy 7)', 
                  fontsize=20) 
title.set_y(1.03) # Move the title a bit for niceness
```
