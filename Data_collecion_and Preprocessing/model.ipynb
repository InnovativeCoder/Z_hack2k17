# Preprocessing the data

```python
from nltk.corpus import stopwords
from nltk.stem import  PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
lemmatizer = WordNetLemmatizer()
stopw = set(stopwords.words("english"))
stopw.update(".", "?", "(", ")", ",", "-", "'", '"',"/","2","1","3","4","5","6","7","8","9","10","&")
```

```python
def text_to_words(raw_text):
    
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    lemmatizer.lemmatize(words)
    words = words.split()
    words = [word for word in words if not word.startswith('rt')]
    meaningful_words = [w for w in words if not w in stopw]
    
    return (" ".join(meaningful_words))
```

```python
final['clean_text'] = final['text'].apply(text_to_words)
final
```

```python
def user_link(raw_text):
    
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    words = words.split()
    words = [word for word in words if word.startswith('@')]
    
    return words
```

```python
def country_name(raw_text):
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    words = words.split()
    for word in words:
        if word == 'spain'or word == 'Portugal' or word =='Italy' or word == 'Turkey' or word == 'Cyprus':
                words = word
        else:
            words = 'Nan'
    return words
```

```python
final['country_name'] = final['clean_text'].apply(country_name)
```

```python
def geturl(raw_text):
    pos_t = (pos_tag(word_tokenize(raw_text)))
    words = raw_text.lower()
    
    words = words.split()
    words = [word for word in words if word.startswith('http')]

    return words
```
